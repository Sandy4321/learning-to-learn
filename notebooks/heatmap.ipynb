```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
```

```python
def plot_heatmap(df, title, size):
    # Plot it out
    fig, ax = plt.subplots()
    heatmap = ax.pcolor(df, cmap=plt.cm.Blues, alpha=0.8)

    # Format
    fig = plt.gcf()
    fig.set_size_inches(size[0], size[1])

    # turn off the frame
    ax.set_frame_on(False)

    # put the major ticks at the middle of each cell
    ax.set_yticks(np.arange(df.shape[0])+0.5, minor=False)
    ax.set_xticks(np.arange(df.shape[1])+0.5, minor=False)

    # set values for x and y axis ticks
    ax.set_xticklabels(df.columns, minor=False, fontsize=15) 
    ax.set_yticklabels(df.index, minor=False, fontsize=15)

    # turn off grid mode
    ax.grid(False)
    
    # set plot title and axis labels
    ax.set_title(title, fontsize=20, y = 1.03)
    ax.set_xlabel('# Categories', fontsize=20)
    ax.set_ylabel('# Exemplars', fontsize=20)
    
    cbar = plt.colorbar(heatmap)
    cbar.set_label('Accuracy', rotation=270, fontsize=20, labelpad=25)
```

```python
df_first1 = pd.read_csv('../results/results_firstOrder1.csv', index_col=0)
df_first2 = pd.read_csv('../results/results_firstOrder2.csv', index_col=0)
df_second1 = pd.read_csv('../results/results_secondOrder1.csv', index_col=0)
df_second2 = pd.read_csv('../results/results_secondOrder2.csv', index_col=0)
```

```python
df_first1
```

```python
plot_heatmap(df_first1, 
             title='First-order Generalization Results (classification accuracy)',
             size=(8, 11))
```

```python
plot_heatmap(df_first2, 
             title='First-order Generalization Results (classification accuracy)',
             size=(4,6))
```

```python
plot_heatmap(df_second1, 
             'Second-order Generalization Results (% of time similarity score correct)',
             size=(8,11))
```
