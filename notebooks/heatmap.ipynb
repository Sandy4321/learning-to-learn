```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../results_secondOrder.csv', 
                 dtype={'nb_categories':np.int16, 
                        'nb_exemplars':np.int16, 
                        'score':np.float32})
```

```python
df_new = pd.DataFrame(index=range(1, 8), columns=range(50, 251, 50))
for i in range(df.shape[0]):
    row = df.iloc[i]
    nb_cat = int(row['nb_categories'])
    nb_ex = int(row['nb_exemplars'])
    score = row['score']
    #print('nb_cat: %i, nb_ex: %i, score: %0.4f' % (nb_cat, nb_ex, score))
    df_new[nb_cat].loc[nb_ex] = score
```

```python
df_new = pd.read_csv('../results_secondOrder.csv', index_col=0)
```

```python
df_new
```

```python
def plot_heatmap(df, title):
    # Plot it out
    fig, ax = plt.subplots()
    heatmap = ax.pcolor(df, cmap=plt.cm.Blues, alpha=0.8)

    # Format
    fig = plt.gcf()
    fig.set_size_inches(8, 11.2)

    # turn off the frame
    ax.set_frame_on(False)

    # put the major ticks at the middle of each cell
    ax.set_yticks(np.arange(df.shape[0])+0.5, minor=False)
    ax.set_xticks(np.arange(df.shape[1])+0.5, minor=False)

    # set values for x and y axis ticks
    ax.set_xticklabels(df.columns, minor=False, fontsize=15) 
    ax.set_yticklabels(df.index, minor=False, fontsize=15)

    # turn off grid mode
    ax.grid(False)
    
    # set plot title and axis labels
    ax.set_title(title, fontsize=20, y = 1.03)
    ax.set_xlabel('# Categories', fontsize=20)
    ax.set_ylabel('# Exemplars', fontsize=20)
    
    plt.colorbar(heatmap)
```

```python
plot_heatmap(df_new, 'Second-order Generalization Results (% of time similarity score correct)')
```

```python
# overwrite original data file
df_new.to_csv('../results_secondOrder.csv')
```
