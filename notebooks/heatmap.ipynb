```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
```

```python
def plot_heatmap(df, title, size, savefile=None):
    # Plot it out
    fig, ax = plt.subplots()
    heatmap = ax.pcolor(df, cmap=plt.cm.Blues, alpha=0.8, vmin=0., vmax=1.)

    # Format
    fig = plt.gcf()
    fig.set_size_inches(size[0], size[1])

    # turn off the frame
    ax.set_frame_on(False)

    # put the major ticks at the middle of each cell
    ax.set_yticks(np.arange(df.shape[0])+0.5, minor=False)
    ax.set_xticks(np.arange(df.shape[1])+0.5, minor=False)

    # set values for x and y axis ticks
    ax.set_xticklabels(df.columns, minor=False, fontsize=15) 
    ax.set_yticklabels(df.index, minor=False, fontsize=15)

    # turn off grid mode
    ax.grid(False)
    
    # set plot title and axis labels
    ax.set_title(title, fontsize=20, y = 1.03)
    ax.set_xlabel('# Categories', fontsize=20)
    ax.set_ylabel('# Exemplars', fontsize=20)
    
    cbar = plt.colorbar(heatmap)
    cbar.set_label('Accuracy', rotation=270, fontsize=20, labelpad=25)
    if savefile is not None:
        plt.savefig(
            savefile, 
            format='pdf',
            bbox_inches='tight'
        )
    else:
        plt.show()
```

## Try doing a contour plot

```python
def contour_plot(df, ex, ca, title, nb_lines=20):
    E, C = np.meshgrid(ex, ca)
    A = np.zeros_like(E, dtype=np.float32)
    for i in range(len(ex)):
        for j in range(len(ca)):
            row = ex[i]
            col = str(ca[j])
            A[j, i] = df[col].loc[row]
    # Plot the contour
    plt.figure(figsize=(10,7))
    CS = plt.contour(C, E, A, levels=np.linspace(0., 1., nb_lines))
    plt.clabel(CS, inline=1, fontsize=10)
    plt.title(title, fontsize=22)
    plt.ylabel('# Exemplars', fontsize=18)
    plt.xlabel('# Categories', fontsize=18)
    plt.show()
```

```python
ex = np.arange(1,10)
ca = np.arange(3,21)
contour_plot(df_first_img, ex, ca, title='First-order Generalization', nb_lines=15)
```

```python
ex = np.arange(1,10)
ca = np.arange(3,21)
contour_plot(df_second_img, ex, ca, title='Second-order Generalization', nb_lines=15)
```
