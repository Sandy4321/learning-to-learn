```python
from __future__ import division, print_function

import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pylab as plt
import itertools
```

```python
df = pd.read_pickle('../data/deepmind_data.pkl')
```

```python
df.head(10)
```

```python
upto = 20000
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18,6))
# First, do the bias plots
for name, marker in zip(df.name.unique(), itertools.cycle('o^+*s')):
    if name == "inet-sl_subj13_50":
        continue
    x = df[df.name==name].time.values
    y = df[df.name==name]['euclidean_linda'].values
    y = y/50.
    
    y = y[np.where(x < upto)]
    x = x[np.where(x < upto)]
    axes[0].plot(x, y, marker=marker, markersize=8)
    axes[0].set_ylim(0,1)
    axes[0].set_xlabel('Epoch', fontsize=12)
    axes[0].set_ylabel('Bias Strength', fontsize=12)
    axes[0].set_title('Shape Bias vs. Training Epoch', fontsize=15)
# Now, do the loss plots
for name, marker in zip(df.name.unique(), itertools.cycle('o^+*s')):
    if name == "inet-sl_subj13_50":
        continue
    x = df[df.name==name].time.values
    y = df[df.name==name]['training_cost'].values
    
    y = y[np.where(x < upto)]
    x = x[np.where(x < upto)]
    axes[1].plot(x, y, marker=marker, markersize=8)
    axes[1].set_ylim(0,1)
    axes[1].set_xlabel('Epoch', fontsize=12)
    axes[1].set_ylabel('Accuracy', fontsize=12)
    axes[1].set_title('Test Set Accuracy vs. Training Epoch', fontsize=15)
```
